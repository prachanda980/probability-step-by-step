## Step 2: Working with NumPy

NumPy is the fundamental package for scientific computing in Python.

```python
import numpy as np
data = np.array([1, 2, 3, 4, 5])
print("Mean:", np.mean(data))
print("Standard Deviation:", np.std(data))
```
