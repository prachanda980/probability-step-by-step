## 1. Arrays and Data Structures

In Python, we don't just use single numbers; we work with collections of data.

### NumPy Arrays
An **array** is a grid of values. They are much faster than standard Python lists for math.

### Pandas DataFrames
A **DataFrame** is like an Excel sheet. It has rows (observations) and columns (features).

```python
# Example: Weight of 5 people in kg
weights = np.array([70, 82, 65, 90, 75])
print("NumPy Array:", weights)

# Example: A small dataset
data = {
    'Age': [25, 30, 35, 40, 28],
    'Salary': [50000, 60000, 75000, 90000, 55000]
}
df = pd.DataFrame(data)
print("\nPandas DataFrame:")
print(df)
```

## 2. Basic Descriptive Statistics

Descriptive statistics summarize the characteristics of a dataset.

### Central Tendency (The Middle)
- **Mean ($\bar{x}$)**: The average. $\bar{x} = \frac{\sum x_i}{n}$
- **Median**: The middle number in a sorted list.
- **Mode**: The most frequent value.

### Dispersion (The Spread)
- **Variance ($\sigma^2$)**: Average of squared differences from the Mean.
- **Standard Deviation ($\sigma$)**: The square root of variance. Most common measure of spread.

```python
scores = np.array([88, 92, 75, 89, 94, 88, 70, 95, 82])

print(f"Mean: {np.mean(scores):.2f}")
print(f"Median: {np.median(scores)}")
print(f"Standard Deviation: {np.std(scores):.2f}")

# Quick visualization of distribution
plt.figure(figsize=(8, 4))
sns.histplot(scores, kde=True, color="skyblue")
plt.title("Distribution of Exam Scores")
plt.show()
```

## 3. Probability Theory & Definitions

**Probability** is the measure of how likely an event is to occur.

**Definition**:
$$P(A) = \frac{\text{Successful Outcomes}}{\text{Total Possible Outcomes}}$$

### Example: Dice Roll
If you roll a 6-sided die, what is the probability of getting a number greater than 4?
- Successes: {5, 6} (Count = 2)
- Total: {1, 2, 3, 4, 5, 6} (Count = 6)
- $P(>4) = 2/6 = 0.33$

```python
def calculate_probability(successes, total):
    return successes / total

prob_dice = calculate_probability(2, 6)
print(f"Probability of rolling > 4: {prob_dice:.2%}")

# Random simulation
rolls = np.random.randint(1, 7, 1000)
success_count = np.count_nonzero(rolls > 4)
print(f"Simulation result (1000 rolls): {success_count/1000:.2%}")
```

## 4. Probability Distributions

A **Probability Distribution** is a mathematical function that describes the probabilities of occurrence of different possible outcomes.

### The Normal Distribution (Bell Curve)
Most natural phenomena (heights, IQ scores) follow this shape. It is defined by two parameters: **Mean (center)** and **Standard Deviation (width)**.

```python
# Generating Normal Distribution data
mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)

plt.figure(figsize=(10, 5))
count, bins, ignored = plt.hist(s, 30, density=True, alpha=0.6, color='g')
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.title("Normal Distribution $(\mu=0, \sigma=0.1)$")
plt.show()
```

## 5. Hypothesis Testing

Hypothesis testing is a way for us to check if our experimental results are "statistically significant" or just happened by chance.

### The T-Test
Used to compare the means of two groups. 
- **Null Hypothesis ($H_0$)**: There is no difference between the groups.
- **Alternative Hypothesis ($H_1$)**: There is a significant difference.
- **P-value**: If p < 0.05, we "reject the null" and say the result is significant.

```python
# Two groups of plant heights (Group A vs Group B)
group_a = np.array([20, 22, 19, 24, 25, 21, 23])
group_b = np.array([28, 30, 27, 26, 29, 31, 28])

t_stat, p_val = stats.ttest_ind(group_a, group_b)

print(f"T-Statistic: {t_stat:.4f}")
print(f"P-Value: {p_val:.6f}")

if p_val < 0.05:
    print("Conclusion: Significant difference found (Reject Null)")
else:
    print("Conclusion: No significant difference found (Fail to reject Null)")

# Visualization
plt.figure(figsize=(8, 5))
sns.boxplot(data=[group_a, group_b], palette="Set2")
plt.xticks([0, 1], ['Group A', 'Group B'])
plt.title("Comparison of Group A and B Heights")
plt.show()
```
