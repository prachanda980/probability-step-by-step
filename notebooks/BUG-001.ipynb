* **Occurred when:** Filtering a Pandas DataFrame to remove outliers using IQR on the `age` column.

## I â€” Isolate
**Conditions under which the bug occurs**

* **Dataset:** `df` (Pandas DataFrame)
* **Column:** `age`
* **Statistical Logic:**
    * $Q1 = 25^{th} \text{ percentile}$
    * $Q3 = 75^{th} \text{ percentile}$
    * $IQR = Q3 - Q1$
    * $\text{Lower Bound} = Q1 - 1.5 \times IQR$
    * $\text{Upper Bound} = Q3 + 1.5 \times IQR$
* **Operation causing the issue:** Attempting to combine two boolean Series using the Python native `and` keyword.

## L â€” Locate
**Exact location of the bug**

* **Notebook cell:** Outlier filtering
* **Faulty line of code:**

```python
filter_data = df[(df['age'] > lower_bound) and (df['age'] < upper_bound)] 



## E â€” Explain
**Why the code fails**

* **Series vs Scalar:** `(df['age'] > lower_bound)` creates a Pandas Series of Booleans.
* **Keyword Limitation:** Python's native `and` operator is designed to evaluate **one** single truth value (a scalar), not a collection (a Series).
* **Ambiguity:** Pandas raises a `ValueError` because it is mathematically ambiguous to evaluate the truth of a whole list of booleans without specifying `.all()` or `.any()`.

---

## R â€” Repair
**How the bug was fixed**

### Fix 1 â€” Use vectorized AND operator (`&`)
The `&` operator is designed for element-wise comparison across entire Pandas Series.

```python
filter_data = df[(df['age'] > lower_bound) & (df['age'] < upper_bound)]
filter_data = df[df['age'].between(lower_bound, upper_bound)]
